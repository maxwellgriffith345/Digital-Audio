```python
#Using MIDO callback
def callBack(event, data=None):
    message, deltatime = event
    print(f"{message} {deltatime}")
```

```python
try:
    midiin, port_name = open_midiinput(2)
    print(port_name)
except (EOFError, KeyboardInterrupt):
    print("that port didn't work")
```

```python
midiin.set_callback(callBack)
```

```python
try:
    # Just wait for keyboard interrupt,
    # everything else is handled via the input callback.
    while True:
        time.sleep(1)
except KeyboardInterrupt:
    print('')
finally:
    print("Exit.")
    midiin.close_port()
    del midiin
```
